```python
#Extract historical weather data from Munich
```

```python
url= 'https://en.tutiempo.net/climate/ws-108660.html'
```

```python
driver=webdriver.Chrome(PATH)
driver.get(url)
```

```python
# Accesing the Title with the path 
```

```python
titulos_= driver.find_element(By.XPATH, '//*[@id="ColumnaIzquierda"]/div/div[6]/table/tbody/tr[1]')
titulos_
```

```python
titular = []

titulo = titulos_.find_elements(By.TAG_NAME, 'th')


for i in range(len(titulo)):
    contenido= titulo[i].text
    titular.append(contenido)
titular
```

```python
#Data of the table
```

```python
tabla = driver.find_element(By.XPATH,'//*[@id="ColumnaIzquierda"]/div/div[6]/table/tbody')

filas = tabla.find_elements(By.TAG_NAME, 'tr')

data=[]

for fila in filas:
    elementos = fila.find_elements(By.TAG_NAME,'td')
    t = []
    for e in elementos:
        t.append(e.text)
    data.append(t)
```

```python
tabla=pd.DataFrame(data, columns=titular)
tabla.head()
```
